```python
import MDAnalysis
from MDAnalysis.analysis.distances import contact_matrix
import glob
import numpy as np
```

```python
af_unrelaxed_pdbs = glob.glob('/gpfs/alpine/world-shared/bif135/alphafold_onsummit/alphafold_test/casp14/af_reduced_db/*/unrelaxed_model_*.pdb')
af_relaxed_pdbs = glob.glob('/gpfs/alpine/world-shared/bif135/alphafold_onsummit/alphafold_test/casp14/af_reduced_db/*/relaxed_model_*.pdb')
omm_gpu_relaxed_pdbs = glob.glob('/gpfs/alpine/world-shared/bif135/alphafold_onsummit/alphafold_test/casp14/af_reduced_db/*/model_*min_00.pdb')
omm_cpu_relaxed_pdbs = glob.glob('/gpfs/alpine/world-shared/bif135/alphafold_onsummit/alphafold_test/casp14/af_reduced_db/*/cpu_relaxation/model_*min_00.pdb')

af_unrelaxed_pdbs.sort()
af_relaxed_pdbs.sort()
omm_gpu_relaxed_pdbs.sort()
omm_cpu_relaxed_pdbs.sort()
```

```python
af_unrelaxed_clash_counts = []
for pdb in af_unrelaxed_pdbs:
    print(pdb)
    u = MDAnalysis.Universe(pdb)
    cas = u.select_atoms('protein and name CA')
    bumps = np.sum(contact_matrix(cas.positions,cutoff = 3.6,returntype='numpy', box=None)) - cas.n_residues
    clashes = np.sum(contact_matrix(cas.positions,cutoff = 1.9,returntype='numpy', box=None)) - cas.n_residues
    af_unrelaxed_clash_counts.append([bumps, clashes])
print(af_unrelaxed_clash_counts)
```

```python
af_relaxed_clash_counts = []
for pdb in af_relaxed_pdbs:
    print(pdb)
    u = MDAnalysis.Universe(pdb)
    cas = u.select_atoms('protein and name CA')
    bumps = np.sum(contact_matrix(cas.positions,cutoff = 3.6,returntype='numpy', box=None)) - cas.n_residues
    clashes = np.sum(contact_matrix(cas.positions,cutoff = 1.9,returntype='numpy', box=None)) - cas.n_residues
    af_relaxed_clash_counts.append([bumps, clashes])
print(af_relaxed_clash_counts)
```

```python
omm_gpu_relaxed_clash_counts = []
for pdb in omm_gpu_relaxed_pdbs:
    print(pdb)
    u = MDAnalysis.Universe(pdb)
    cas = u.select_atoms('protein and name CA')
    bumps = np.sum(contact_matrix(cas.positions,cutoff = 3.6,returntype='numpy', box=None)) - cas.n_residues
    clashes = np.sum(contact_matrix(cas.positions,cutoff = 1.9,returntype='numpy', box=None)) - cas.n_residues
    omm_gpu_relaxed_clash_counts.append([bumps, clashes])
print(omm_gpu_relaxed_clash_counts)
```

```python
omm_cpu_relaxed_clash_counts = []
for pdb in omm_cpu_relaxed_pdbs:
    print(pdb)
    u = MDAnalysis.Universe(pdb)
    cas = u.select_atoms('protein and name CA')
    bumps = np.sum(contact_matrix(cas.positions,cutoff = 3.6,returntype='numpy', box=None)) - cas.n_residues
    clashes = np.sum(contact_matrix(cas.positions,cutoff = 1.9,returntype='numpy', box=None)) - cas.n_residues
    omm_cpu_relaxed_clash_counts.append([bumps, clashes])
print(omm_cpu_relaxed_clash_counts)
```

```python
af_unrelaxed_clash_counts = np.array(af_unrelaxed_clash_counts)
af_relaxed_clash_counts   = np.array(af_relaxed_clash_counts)
omm_gpu_relaxed_clash_counts = np.array(omm_gpu_relaxed_clash_counts)
omm_cpu_relaxed_clash_counts = np.array(omm_cpu_relaxed_clash_counts)

print(np.mean(af_unrelaxed_clash_counts[:,0]),np.std(af_unrelaxed_clash_counts[:,0]),np.max(af_unrelaxed_clash_counts[:,0]))
print(np.mean(af_relaxed_clash_counts[:,0]),np.std(af_relaxed_clash_counts[:,0]),np.max(af_relaxed_clash_counts[:,0]))
print(np.mean(omm_gpu_relaxed_clash_counts[:,0]),np.std(omm_gpu_relaxed_clash_counts[:,0]),np.max(omm_gpu_relaxed_clash_counts[:,0]))
print(np.mean(omm_cpu_relaxed_clash_counts[:,0]),np.std(omm_cpu_relaxed_clash_counts[:,0]),np.max(omm_cpu_relaxed_clash_counts[:,0]))

print(np.mean(af_unrelaxed_clash_counts[:,1]),np.std(af_unrelaxed_clash_counts[:,1]),np.max(af_unrelaxed_clash_counts[:,1]))
print(np.mean(af_relaxed_clash_counts[:,1]),np.std(af_relaxed_clash_counts[:,1]),np.max(af_relaxed_clash_counts[:,1]))
print(np.mean(omm_gpu_relaxed_clash_counts[:,1]),np.std(omm_gpu_relaxed_clash_counts[:,1]),np.max(omm_gpu_relaxed_clash_counts[:,1]))
print(np.mean(omm_cpu_relaxed_clash_counts[:,1]),np.std(omm_cpu_relaxed_clash_counts[:,1]),np.max(omm_cpu_relaxed_clash_counts[:,1]))
```
